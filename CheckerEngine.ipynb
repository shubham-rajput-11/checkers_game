```python
import sys
import pygame as p
import CheckerEngine
WIDTH = HEIGHT = 512
DIMENSION = 8
SQ_SIZE = HEIGHT // DIMENSION
MAX_FPS = 15  # for animation later
IMAGES = {}

def loadImages():
    IMAGES['nw']=p.transform.scale(p.image.load('image/nw.png'),(SQ_SIZE,SQ_SIZE))
    IMAGES['nb']=p.transform.scale(p.image.load('image/nb.png'),(SQ_SIZE,SQ_SIZE))
    IMAGES['kw']=p.transform.scale(p.image.load('image/kw.png'),(SQ_SIZE,SQ_SIZE))
    IMAGES['kb']=p.transform.scale(p.image.load('image/kb.png'),(SQ_SIZE,SQ_SIZE))
    
# we can put images of any piece by IMAGE[piece]
    
def main():
    p.init()
    screen = p.display.set_mode((WIDTH, HEIGHT))
    clock = p.time.Clock()
    screen.fill(p.Color("white"))
    gs = CheckerEngine.GameState()
#     gs = GameState()
    validMoves = gs.getValidMoves()   # generating valid moves   expensive operation that why flag variable
    moveMade = False     # flag variable when a move is made

    loadImages()    # only once
    running = True
    sqSelected = () # no square selected initially , keep track of last click of the user ( tuple:row,col)
    playerClicks = [] # keep track of players clicks ( 2 tuple : [(6,7),(4,4)])
    
    gameOver=False
    
    while running:
        for e in p.event.get():
#             if e.type == p.quit():
            if e.type == p.QUIT:
                running = False
                p.quit()
                sys.exit()
                
                
                
            # mouse event handler
            
            elif e.type == p.MOUSEBUTTONDOWN:
                if not gameOver:
                    location = p.mouse.get_pos()   # returns (x,y) location of mouse in screen
                    row = location[1]//SQ_SIZE
                    col = location[0]//SQ_SIZE

                    if sqSelected ==(row,col):  # user click same square twice 
                        sqSelected = ()   #clear selected square
                        playerClicks = [] # clear player clicks
                    else:
                        sqSelected = (row,col)
                        playerClicks.append(sqSelected)   #append for both 1st ans 2nd click

                    if len(playerClicks) == 2 :  # after 2nd click
                        move = CheckerEngine.Move(playerClicks[0],playerClicks[1], gs.board)  #player generated move
                        print(move.endRow +1,move.endCol+1)
                        for i in range(len(validMoves)):
                            if move == validMoves[i]:
                                gs.makeMove(validMoves[i])    # move variable here is actually variable from above line 
                                moveMade = True     # thus move is made so we need to generate new set of valid moves
                                sqSelected = ()
                                playerClicks = []        # move is made then sqSelected and playerClicks to be reset to empty
                        if not moveMade:
                            playerClicks = [sqSelected]
                
            # key handlers
            
            elif e.type == p.KEYDOWN:
                if e.key == p.K_z :   #undo when z is pressed
                    gs.undoMove()
                    moveMade = True    # calling getvaid moves one time 
                    
                if e.key == p.K_r: # reset the board when 'r' is pressed
                    gs=CheckerEngine.GameState()
                    validMoves = gs.getValidMoves()
                    sqSelected = ()
                    playerClicks = []
                    movemade = False
                    
                    
        if moveMade:
            validMoves = gs.getValidMoves()
            moveMade = False
                        
        drawGameState(screen,gs)
        
        if gs.checkMate():
            gameOver = True
            if gs.blackToMove:
                drawText(screen,"White wins!!")
            else:
                drawText(screen,"black wins!!")
        
        clock.tick(MAX_FPS)
        p.display.flip()
        
#error here module object is not callable
def drawText(screen , text):
    font=p.font.SysFont("Helvitca",32,True,False)
    textObject= font.render(text,0,p.Color('Black'))
    textLocation = p.Rect(0,0,WIDTH,HEIGHT).move(WIDTH/2 - textObject.get_width()/2 , HEIGHT/2 -textObject.get_height()/2)
    screen.blit(textObject , textLocation)
        
def drawGameState(screen,gs):
    drawBoard(screen) # draw squares on the board

    drawPieces(screen, gs.board)  # draw pieces on top of those squares

'''draw the squres on the board '''
def drawBoard(screen):
    colors = [p.Color("white"),p.Color("dark green")]
    for row in range(DIMENSION):
        for col in range(DIMENSION):
            color = colors[((row + col)% 2)]
            p.draw.rect(screen,color,p.Rect( col*SQ_SIZE, row*SQ_SIZE, SQ_SIZE,  SQ_SIZE))    # col is x axis     row is y axis

def drawPieces(screen, board):
    for row in range(DIMENSION):
        for col in range(DIMENSION):
            piece = board[row][col];
            if piece != "--": # not empty square
                screen.blit(IMAGES[piece], p.Rect(col*SQ_SIZE,row*SQ_SIZE,SQ_SIZE,SQ_SIZE))
            
            
            
if __name__=="__main__":
    main()
```
